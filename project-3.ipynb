### Получения страны из адреса

```python
def country_feature(hotel_address):
    hotel_address = str(hotel_address)
    
    address_lst = hotel_address.split()
    
    last_word = address_lst[-1]
    
    if last_word == 'Kingdom':
        return 'United Kingdom'
    
    return last_word
```

### Получение города из адреса

```python
def city_feature(hotel_address):
    hotel_address = str(hotel_address)
    
    address_lst = hotel_address.split()
    
    last_word = address_lst[-1]
    
    if last_word == 'Kingdom':
        return address_lst[-5]
    
    return address_lst[-2]
```

### Расстояние до центра города

```python
# Исходные признаки lat и lng собираю в один в виде списка из двух координат
full_data['loc'] = list(full_data[['lat', 'lng']].to_records(index=False))
```

```python
# Функция для определения расстояния по координатам до центров городов
def get_dist_to_center(cols):
    loc = cols[0]
    city = str(cols[1])
    
    # Данные взял из гугла
    city_center = {
    'London': (51.5072, 0.1276),
    'Barcelona': (41.3874, 2.1686),
    'Paris': (48.8566, 2.3522),
    'Amsterdam': (52.3676, 4.9041),
    'Vienna': (48.2082, 16.3738),
    'Milan': (45.4642, 9.1900)
    }
    
    loc_city_center = city_center[city]
    
    return hs.haversine(loc, loc_city_center)
```

```python
full_data['dist_to_center'] = full_data[['loc', 'cat_city']].apply(get_dist_to_center, axis=1)
```

```python
def get_tags_list(tags_string):
    tags_string = str(tags_string)
    
    tags_string = tags_string.replace('[', '').replace(']', '').replace('\'', '')
    tag_list = tags_string.split(',')
    
    # Удаляю из элементов списка лишние пробелы по сторонам
    # Помещаю эелементы в сет, что бы измежать потенциальных дубликатов тегов
    tag_set = set([i.strip() for i in tag_list])
    
    # Распаковываю назад в список
    tag_list = list(tag_set)
    
    tag_list.sort() # сортирую для удобства в алфавитном порядке
    
    return tag_list
```

```python
full_data['tags'] = full_data['tags'].apply(get_tags_list)
```

### Сколько уникальных тегов представлено в наборе данных?

### rooms признак количество комнат в номере

```python
def get_rooms_feature(tags):
    # обхожу список тегов
    for tag in tags:
        # если в теге есть слово
        if 'rooms' in tag:
            num_rooms = tag.split()[0]
            
            if num_rooms.lower() == 'two':
                return 2
            elif num_rooms.lower() == 'three':
                return 2
            
            try:
                return int(num_rooms)
            
            except ValueError as e:
                return 1
    
    return 1 # Если обойдя список тегов не нашел тег с количеством комнат, то считаю что это сттандартный однокомнатный номер
```

### Анализ тональности

```python
sia = SentimentIntensityAnalyzer()

pos_pol_scores = full_data['positive_review'].apply(lambda x: sia.polarity_scores(x))
neg_pol_scores = full_data['negative_review'].apply(lambda x: sia.polarity_scores(x))

full_data['pos_neu_sentiment'] = pos_pol_scores.apply(lambda x: abs(x['neu']))
full_data['neg_neu_sentiment'] = neg_pol_scores.apply(lambda x: abs(x['neu']))
full_data['neu_sentiment'] = full_data['pos_neu_sentiment'] - full_data['neg_neu_sentiment']

full_data['pos_sentiment'] = pos_pol_scores.apply(lambda x: abs(x['compound']))
full_data['neg_sentiment'] = neg_pol_scores.apply(lambda x: abs(x['compound']))
full_data['sentiment'] = full_data['pos_sentiment'] - full_data['neg_sentiment']

full_data['pos1_sentiment'] = pos_pol_scores.apply(lambda x: abs(x['pos']))
full_data['pos2_sentiment'] = neg_pol_scores.apply(lambda x: abs(x['pos']))

full_data['neg1_sentiment'] = neg_pol_scores.apply(lambda x: abs(x['neg']))
full_data['neg2_sentiment'] = pos_pol_scores.apply(lambda x: abs(x['neg']))

full_data['full_sentiment'] = full_data['pos1_sentiment'] + full_data['pos2_sentiment'] - full_data['neg1_sentiment'] - full_data['neg2_sentiment']
```
